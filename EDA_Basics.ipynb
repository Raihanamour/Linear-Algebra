### GroupBy

1. can be applied on categorical variable
2. Group data into categories
3. single or multiple variables 

```python
df.head()
```

```python
new_df = df[['Company','Ram', 'Price']]
```

```python
df_group = new_df.groupby(['Company','Ram']).mean()
```

```python
df_group.head(10)
```

```python
df_group = new_df.groupby(['Company','Ram'],as_index=False).mean()
df_group.head(10)
```

### Pivot()

One variable displayed along col and other variable along rows

```python
df.head()
```

```python

df_pivot = df_group.pivot(index='Company',columns='Ram')
```

```python
df_pivot.head(10)
```

### heatmap 
plot target variable over multiple variables

```python
plt.pcolor(df_pivot,cmap='RdBu')
plt.colorbar()
plt.show()
```
