## 4. Binning

```python
Sal = pd.read_csv('/Users/raihan/Desktop/Dataset/Salary Data.csv')
```

```python
Sal.head()
```

```python
bins = np.linspace(min(Sal['Salary']),max(Sal['Salary']),4)
```

```python
bin_labels = ['low','medium','high']
```

```python
Sal['Salary_bin'] = pd.cut(Sal['Salary'],bins = bins, labels= bin_labels, right = False)
```
